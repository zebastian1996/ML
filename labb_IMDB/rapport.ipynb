Introduktion                                                                
                
    Målet med programet var att generera filmrekomendationer baserat på en vald film.
    
· Data-analys (EDA)
    
    Resultatet visar att det finns samband mellan användare som ger högt betyg till samma filmer.
    Användarna brukar tendera att tycka om liknande filmer 

    genrer har dessutom visat sig har en stark koppling mellan liknande filmer så detta kontrolleras i slutet
    
· Modell

    Rekomendationerna är baserat på användare som ger den valda film minst 4. Vi kollar där efter
    vidare på vad med dom har gillat för att identifiera relativa rekommendationer
    
    Resultatet divideras och där efter tar vi roten ur resultatet för att minimera risken att generellt
    populära filmer alltid blir rekomenderade 
    
    Slutligen kollar programet om det finns likheter i genrer och ger en extra boost till filmer 
    som machar med genrer mer en andra
